### Initialize punctuations and structures

```python
unique_words_dict = SortedDict({})
digits = [str(x) for x in range(10)]
punctuations = list(string.punctuation) + ['_', '\t', '\n']
puncs = ["'"]
punctuations.remove("'")
total_sentences = 0
print(punctuations)
```

### Function to process the sentences and save them in a file

```python
def process_review(review):
    global total_sentences
    sentences = review.split('.')  ## split sentences using full-stop
    processed_sentences = []
    for sentence in sentences:
        if not sentence :
            continue
        rpun = '[' + re.escape(''.join(punctuations)) + ']'
        rp = '[' + re.escape(''.join(puncs)) + ']'
        sentence = re.sub(rpun, ' ', sentence.strip())  ## remove punctuations/symbols and split words using them
        sentence = re.sub(rp, '', sentence.strip())  ## join letters after apostrophe with the original word : I've --> Ive
        sentence = re.sub(r'\w*\d\w*', '', sentence).strip()  ## omit words having digits in them
        words = sentence.strip().lower().split()  ## casefolding sentence and splitting into words
        if len(words) <= 1:  ## omit if sentence has only 1 word
            continue
        for word in words:
            count = unique_words_dict.get(word,0)  ## increment count of the word
            count += 1
            unique_words_dict[word] = count
        sentence = ' '.join(words)
        total_sentences += 1
        processed_sentences.append(sentence)
    return processed_sentences

def write_stats():
    with open('unique_words.json', 'w') as fa:
        fa.write(json.dumps(unique_words_dict))
    with open('meta.txt', 'w') as fb:
        fb.write(str(total_sentences))
```

### Load the corpus and extract the review and preprocess it

```python
with open('processed_reviews.txt', 'w') as f2:
    pass
i = 0
with open("reviews_Electronics_5.json") as f1:
    with open('processed_reviews.txt', 'a') as f2:
        for line in f1:
            line_json = json.loads(line.strip())
            # print(line_json['reviewText'])
            sentences = process_review(line_json['reviewText'])
            for s in sentences:
                f2.write(s+"\n")
            print(i)
            i += 1
print(len(unique_words_dict))
write_stats()
```
